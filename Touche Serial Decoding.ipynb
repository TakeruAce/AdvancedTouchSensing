```python
import serial
import time

portName = '/dev/tty.usbmodem1451'


def main():
    
    startTime = time.time()
    
    # Grobal Variable
    xValue = 0
    yValue = 0
    Command = 0
    Error = False
    UpdateGraph = True
    lineGraph = 0
    ErrorCounter = 0
    TotalReceived = 0

    # local variable

    DataReceived1 = False
    DataReceived2 = False
    DataReceived3 = False

    DynamicArrayTime1 = []
    DynamicArrayTime2 = []
    DynamicArrayTime3 = []

    Time1 = []
    TIme2 = []
    Time3 = []

    Voltage1 = []
    Voltage2 = []
    Voltage3 = []

    current = []

    DynamicArray1 = []
    DynamicArray2 = []
    DynamicArray3 = []

    PowerArray = []
    DynamicArrayPower = []
    DynamicArrayTime = []

    data1Received=False
    data2Received=False
    data3Received=False

    increment = 0

    NumOfSerialBytes = 8
    serialInArray = [0] * NumOfSerialBytes
    serialCount = 0

    xMSB = 0
    xLSB = 0
    yMSB = 0
    yLSB = 0

    
    with serial.Serial(portName,115200,timeout=1) as ser:
        while True:
            
#             一定時間立ったら終了
            currentTime = time.time()
            if currentTime - startTime > 1 :
                break
            
#             check the value read from serial is good
#             bytes型取得なのでintに変換
            inByte = ser.read()
            inByte = int.from_bytes(inByte,'big')
#             print(inByte)
            
            if inByte == 0 :
                serialCount = 0
        
            if (inByte > 255) :
                println(" inByte = " + inByte)
                break
                
                
#             add latest byte from the serial to array
            serialInArray[serialCount] = inByte
            serialCount += 1
        
            Error = True
            if serialCount >= NumOfSerialBytes :
                serialCount = 0
                
                TotalReceived += 1
                
                checkSum = 0
                
                for i in range(0,len(serialInArray) - 1) : 
                    checkSum = checkSum+serialInArray[i]
                
                checkSum = checkSum%255
            
                if checkSum==serialInArray[7] :
                    Error = False
                    DataReceived = True
                else :
                    Error = True
                    DataReceived=False
                    ErrorCounter += 1
                    print("checkSum : %d" % checkSum)
                    print("Error :" + str(ErrorCounter) + "/" + str(TotalReceived) + " : " + str(float(ErrorCounter/TotalReceived)*100)+"%")
            
            if Error == False :
                zeroByte = serialInArray[6]
                
                xLSB = serialInArray[3]
                if (zeroByte & 1) == 1 :
                    xLSB = 0
                    
                xMSB = serialInArray[2]
                if (zeroByte & 2) == 2 :
                    xMSB = 0
                    
                yLSB = serialInArray[5]
                if (zeroByte & 4) == 4 :
                    yLSB = 0
                    
                yMSB = serialInArray[4]
                if (zeroByte & 8) == 8 :
                    yMSB = 0
                    
                Command = serialInArray[1]
                
                xValue = xMSB << 8 | xLSB
                yValue = yMSB << 8 | xLSB
                
                if Command == 1 :
                    DynamicArrayTime3.append(xValue)
                    DynamicArray3.append(yValue)
                elif Command == 2 :
                    DynamicArrayTime3 = []
                    DynamicArray3 = []
                elif Command == 3 :
                    Time3 = DynamicArrayTime3
                    Voltage3 = DynamicArray3
                    DataReceived3 = True
                elif Command == 4 :
                    DynamicArrayTime2.append(xValue)
                    DynamicArray2.append(yValue)
                elif Command == 5 :
                    DynamicArray2 = []
                    DynamicArrayTime2 = []
                elif Command == 6 :
                    Time2 = DynamicArrayTime2;
                    current=DynamicArray2
                    DataReceived2 = True
                elif Command == 20 :
                    PowerArray.append(yValue)
                elif Command == 21 :
                    DynamicArrayTime.append(xValue)
                    DynamicArrayPower.append(yValue)
            
            if Voltage3 : 
                print(Voltage3)
        ser.close()
```
